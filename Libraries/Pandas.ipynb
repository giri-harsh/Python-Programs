## Cell 1.4 : Filling & Dropping Nulls

```python
print("Before:", df.isnull().sum())
```

* printing before cleaning so that we can compare

```python
df['Age'] = df['Age'].fillna(
    df['Age'].median()
)
```

* `.fillna()` replaces all NaN values with something you choose
 * we fill with the `MEDIAN` age — more robust than mean

```python
df = df.drop(columns=['Cabin'])
```

* `.drop(columns=[...])` removes entire columns from the DataFrame
* we drop `Cabin` because `77%` of its values are missing
* keeping it would confuse any model we train later

```python
df['Embarked'] = df['Embarked'].fillna(   # fill the 2 missing Embarked values
    df['Embarked'].mode()[0]              # .mode() returns the most common value (S, C, or Q)
)
```

* fill the 2 missing Embarked values
* `.mode()` returns most common values ( S,C,or Q)

```python
print("After:", df.isnull().sum())
```

* print missing counts AFTER cleaning to confirm everything is fixed

## 1.5 Selecting and Filtering

```python
df['Survived']
```

* selecting a single column returns a Series (1D)
* 0 = did not survive, 1 = survived

```python
df[['Name', 'Age', 'Survived']]
```

* selecting MULTIPLE columns using a list inside []
* always use double brackets `[[...]]` for multiple columns

```python
survivors = df[df['Survived'] == 1]
print(survivors.shape)
```

* boolean filtering: keep only rows where Survived equals 1
* `df['Survived'] == 1` creates a `True/False` Series
 * the outer `df[...]` keeps only the True rows
